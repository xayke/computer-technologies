### **Matplotlib**
**Matplotlib** - это библиотека Python для построения 2D графиков.  
Процесс построение графика состоит из 6 шагов.  


####**1. Подготовка данных**

```python
import numpy as np

x = np.linspace(0, 10, 100)
y = np.cos(x)
z = np.sin(x)
```

#### **2. Создание графика**

```python
import matplotlib.pyplot as plt

figure_1 = plt.figure()
figure_2 = plt.figure(figsize=plt.figaspect(2.0))

figure_1.add_axes()
axes_1 = figure_1.add_subplot(221)
axes_2 = figure_1.add_subplot(212)
figure_3, axes_3 = plt.subplots(nrows=2, ncols=2)
figure_4, axes_4 = plt.subplots(ncols=3)
```

#### **3. Отрисовка графика**  

##### **3.1 Рисуем фигуры**

###### *3.1.1 Отрисовка точек, соединенных линиями*

```python
figure, ax = plt.subplots()
lines = ax.plot(x, y)
```

###### *3.1.2 Отрисовка точек без соединительных элементов*

```python
points_figure, points_ax = plt.subplots()
points = points_ax.scatter(x, y)
```

###### *3.1.3 Отрисовка гистограмм и прямых линий*

```python
axes_3[0,0].bar([1,2,3], [3,4,5]) # Левый верхний график: гистограмма
axes_3[1,0].barh([0.5,1,2.5], [0,1,2]) # Левый нижний график: горизонтальная гистограмма
axes_3[1,1].axhline(0.45) # Праваый нижний график: горизонтальная линия
axes_3[0,1].axvline(0.65) # Праваый верхний график: вертикальная линия
figure_3
```

###### *3.1.4 Закрашивание фигуры и её разницы по оси y*

```python
ax.fill(x, y, color='blue')
ax.fill_between(x,y,color='yellow')
figure
```

##### **3.2 Распределения данных**

```python
axes_1.hist(y)
axes_2.boxplot(y)
axes_2.violinplot(z)
figure_1
```

#### **4. Настройка графика**


Обычно это обычная кастомизация графика:
1. Подписи осей;
2. Выбор цвета фигур;
3. Настройка предельных значений и делений осей;
4. Настройка стиля линии, которой рисуются фигуры и др.

#### **5. Сохранение результата**

Результат можно сохранить в файл в виде изображения:

```python
figure.savefig('foo.png')
figure.savefig('foo2.png', transparent=True) #вариант с прозрачными фигурами
```

#### **6. Отображение графика**

```python
plt.show()
```

### **Seaborn**

**Seaborn** - библиотека Python основанная на библиотеке matplotlib и предлагающая работу с более красивыми графиками. 
Например так можно быстро построить график, настроив подписи к полям и его название.

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
sns.set_style('whitegrid')
g = sns.lmplot(x='tip',
               y='total_bill',
               data=tips,
               aspect=2)
g = (g.set_axis_labels('Tip', 'Total bill (USD)')).set(xlim=(0, 10), ylim=(0,100))
plt.title('Title')
plt.show(g)
```

Основная цель Seaborn состоит в том, чтобы быстро и красиво вывести информацию с разных источников в красивом виде, которую можно будет применять для анализа данных. В целом в этой библиотеке сохраняется функционал matplotlib, но с более удобными и гибкими настройками.

### **Bokeh**

**Bokeh** - это библиотека Python для интерактивной визуализации данных, позволяющая возможность отобразить в удобном виде большие наборы данных в web-браузерах.  
Для создания графика требуется пройти 5 этапов:
1. Подготовка данных
2. Определяем график
3. Выбрать отрисовку для данных
4. Выбор средств отображения результата
5. Отображение или сохранение результата

```python
from bokeh.plotting import figure
from bokeh.io import output_file, show

# шаг 1 - подготовка данных
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
# шаг 2 - определяем график
p = figure(title='simple line example', 
           x_axis_label='x',
           y_axis_label='y')
# шаг 3 - выбираем отрисовку для данных
p.line(x, y, legend_label='Temp.', line_width=2)
# шаг 4 - выбираем средство отображение результата
output_file('lines.html')
# шаг 5 - отображение результата
show(p)
```
